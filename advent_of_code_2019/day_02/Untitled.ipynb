```python
s = list(map(lambda x: int(x), s.split(',')))
print(s)
```

```python
t1 = [1,1,1,4,99,5,6,0,99]
t2 = [2,4,4,5,99,0]
```

```python
def process(s):
    x = s[:]
    for i in range(0, len(x), 4):
        opcode = x[i]
        if opcode == 99:
            # quit
            break
        p1 = x[x[i + 1]]
        p2 = x[x[i + 2]]
        loc = x[i + 3]
        if opcode == 1:
            # add
            x[loc] = p1 + p2
            continue
        if opcode == 2:
            # multiply
            x[loc] = p1 * p2
            continue
        # should never reach
        print('error')
        break
    return x
```

```python
print(process(t1))
print(process(t2))
```

```python
print(process(tmp))
done = process(tmp)
```
